```python
import json
import os
```

```python
file_root = '/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_all/Session_all/individual_summarization/llama3.1:8b'
output_root = '/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_all/Session_all/individual_summarization/llama3_combined'

for file in os.listdir(file_root):
    if file.endswith('.json'):
        file_name = file.split('&')[0]
        combined_data = {}
        for file in os.listdir(file_root):
            if file.endswith('.json') and file.split('&')[0] == file_name:
                speaker_id = file.split('&')[1].split('.')[0]
                with open(os.path.join(file_root, file), 'r') as f:
                    data_part = json.load(f)
                    combined_data[speaker_id] = data_part["summary"].split(": ")[1]

            output_file = os.path.join(output_root, file_name + '.json')
            with open(output_file, 'w') as f:
                json.dump(combined_data, f, indent=4)





            
```

```python
final_output_root = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/project/annotation/annotation_json"

speaker_path = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_all/Session_all/individual_summarization/llama3_combined"
topic_path = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_all/Session_all/summarization/8b_e+t+"
output_path = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/project/annotation/summary1"

for file in os.listdir(topic_path):
    if not file.endswith('.json'):
        continue 
    topics = {} 
    with open(os.path.join(topic_path, file), 'r') as f:
        topic_data = json.load(f)
       
    for topic, content in topic_data.items():
        topics[topic] = content.get("summary", "").strip()
    with open(os.path.join(speaker_path, file), 'r') as f:
        speaker_data = json.load(f) 
    with open(os.path.join(output_path, file), 'r') as f:
        output_data = json.load(f)
    
    fianl_output_file = os.path.join(final_output_root, file)
    with open(fianl_output_file, 'w') as f:
        json.dump({
            "topic": topics,
            "speaker": speaker_data,
            "summary": output_data["summary"].split(": ")[1].strip()
        }, f, indent=4)
```
