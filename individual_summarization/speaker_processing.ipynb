### This code is to extract individual speaker transcript for speaker level summarization

```python
import os
import json

transcript_path = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/ami/transcript_json_filtered_emotion"
out_path = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/ami/individual_transcript_mapping"
os.makedirs(out_path, exist_ok=True)
```

```python
for file in os.listdir(transcript_path):
    if file.endswith(".json"):
        with open(os.path.join(transcript_path, file), 'r') as f:
            data = json.load(f)
            speakers = {entry["speaker"] for entry in data.values()}
        for speaker in speakers:
            out_path_file_path = os.path.join(out_path, file.replace(".json", f"&{speaker}.json"))
            transcript = {}
            index = 0
            for k,v in data.items():
                if v["speaker"] == speaker:

                    transcript[k] = index
                    index += 1


            with open(out_path_file_path, 'w') as out_f:
                json.dump(transcript, out_f, indent=4)
```
