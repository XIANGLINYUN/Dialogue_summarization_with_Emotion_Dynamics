```python
import os
import json
import re
from llama3_tokenizer import Tokenizer

TOKENIZER_MODEL_PATH = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/Meta-Llama-3-8B-Instruct/original/tokenizer.model"  # <-- CHANGE THIS
tokenizer = Tokenizer(TOKENIZER_MODEL_PATH)

def count_llama_tokens(text: str) -> int:
    # No BOS/EOS for length; you can set bos/eos True if you want model-input length.
    return len(tokenizer.encode(text, bos=False, eos=False))
```

```python
def count_words(text):
    # Counts words, ignores punctuation, keeps contractions as one word
    return len(re.findall(r"\b\w+(?:'\w+)?\b", text))
```

```python
UTTERANCE_FILE_ROOT = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_all/Session_all/transcript_json"

ALL_UTTERANCE_COUNT = 0
ALL_WORD_COUNT = 0
ALL_TOKEN_COUNT = 0

for file in os.listdir(UTTERANCE_FILE_ROOT):
    if file.endswith(".json"):
        with open(os.path.join(UTTERANCE_FILE_ROOT, file), 'r') as f:
            data = json.load(f)
            for key, utterance in data.items():
                ALL_WORD_COUNT += count_words(utterance)
                ALL_TOKEN_COUNT += count_llama_tokens(utterance)
                
        
        ALL_UTTERANCE_COUNT += len(data)
print(f"Total Utterances: {ALL_UTTERANCE_COUNT}")
print(f"Total Words: {ALL_WORD_COUNT}")

avg_words_per_utterance = ALL_WORD_COUNT / ALL_UTTERANCE_COUNT if ALL_UTTERANCE_COUNT > 0 else 0
print(f"Average Words per Utterance: {avg_words_per_utterance:.2f}")
print(f"Total LLaMA Tokens: {ALL_TOKEN_COUNT}")
avg_tokens_per_utterance = ALL_TOKEN_COUNT / ALL_UTTERANCE_COUNT if ALL_UTTERANCE_COUNT > 0 else 0
print(f"Average LLaMA Tokens per Utterance: {avg_tokens_per_utterance:.2f}")
        
```

```python
DURANTION_FILE_ROOT = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_CHUNK/utterance_json"

TOTAL_DURATION_UTTEREANCE = 0

for dir in os.listdir(DURANTION_FILE_ROOT):
    dir_path = os.path.join(DURANTION_FILE_ROOT, dir)
    for file in os.listdir(dir_path):
        if file.endswith(".json"):
            with open(os.path.join(dir_path, file), 'r') as f:
                data = json.load(f)
                for key,segment in data.items():
                    start_time = segment['start_time']
                    end_time = segment['end_time']
                    TOTAL_DURATION_UTTEREANCE += (end_time - start_time)


avg_duration_per_utterance = TOTAL_DURATION_UTTEREANCE / ALL_UTTERANCE_COUNT if ALL_UTTERANCE_COUNT > 0 else 0
print(f"Total Duration of all Utterances (seconds): {TOTAL_DURATION_UTTEREANCE:.2f}")
print(f"Total Utterances for Duration Calculation: {ALL_UTTERANCE_COUNT}")
print(f"Average Duration per Utterance (seconds): {avg_duration_per_utterance:.2f}")    
```

```python
TOTAL_DURATION_TOPIC = 0
TOTOAL_WORDS_TOPIC = 0
TOTAL_TOKENS_TOPIC = 0
TOTAL_TOPICS = 0

for dir in os.listdir(DURANTION_FILE_ROOT):
    dir_path = os.path.join(DURANTION_FILE_ROOT, dir)
    for file in os.listdir(dir_path):
        if file.endswith(".json"):
            print(f"Processing file: {file}")
            topic_file = os.path.join("/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/IEMOCAP_all/IEMOCAP_70b_segmented", file)
            with open(os.path.join(dir_path, file), 'r') as f:
                data = json.load(f)
            with open(topic_file, 'r') as f:
                topic_data = json.load(f)
            for topic_info in topic_data:
                start_time = data[str(topic_info["start_sentence"])]["start_time"]
                end_time = data[str(topic_info["end_sentence"])]["end_time"]
                for i in range(topic_info["start_sentence"], topic_info["end_sentence"] + 1):
                    utterance = data[str(i)]["text"]
                    TOTOAL_WORDS_TOPIC += count_words(utterance)
                    TOTAL_TOKENS_TOPIC += count_llama_tokens(utterance)
                TOTAL_DURATION_TOPIC += (end_time - start_time)
                TOTAL_TOPICS += 1

print(f"Total Duration of all Topics (seconds): {TOTAL_DURATION_TOPIC:.2f}")
avg_duration_per_topic = TOTAL_DURATION_TOPIC / TOTAL_TOPICS if TOTAL_TOPICS > 0 else 0
print(f"Total Topics for Duration Calculation: {TOTAL_TOPICS}") 
print(f"Average Duration per Topic (seconds): {avg_duration_per_topic:.2f}")

avg_words_per_topic = TOTOAL_WORDS_TOPIC / TOTAL_TOPICS if TOTAL_TOPICS > 0 else 0
print(f"Total Words in all Topics: {TOTOAL_WORDS_TOPIC}")               
print(f"Average Words per Topic: {avg_words_per_topic:.2f}")
avg_tokens_per_topic = TOTAL_TOKENS_TOPIC / TOTAL_TOPICS if TOTAL_TOPICS > 0 else 0
print(f"Total LLaMA Tokens in all Topics: {TOTAL_TOKENS_TOPIC}")               
print(f"Average LLaMA Tokens per Topic: {avg_tokens_per_topic:.2f}")    
            
                
```
