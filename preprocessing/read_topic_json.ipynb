```python
import os
import json

raw_data_path = "/Users/linyunxiang/NL/TUD/Project/2025/dynamic_summary/dataset/ami/ami_70b_segmented"

for file in os.listdir(raw_data_path):
    if file.endswith(".txt"):
        file_path = os.path.join(raw_data_path, file)
        out_path = os.path.join(raw_data_path, file.split(".")[0] + ".json")

        with open(file_path, 'r') as f:
            data = json.load(f)

        # This is a STRING that contains JSON
        content = data["message"]["content"]

        # Parse the inner JSON
        parsed_content = json.loads(content)

        # Write proper JSON
        with open(out_path, 'w') as out_f:
            json.dump(parsed_content, out_f, indent=4)

        print(f"Saved cleaned JSON to {out_path}")
```
